```python
from transformers import AutoTokenizer

# Carregar tokenizer do GPT-2 (mesmo tipo usado nos LLMs modernos)
tokenizer = AutoTokenizer.from_pretrained("gpt2")
print("Tokenizer carregado!")
```

```python
frases_en = [
    "The cat sat on the mat.",
    "Artificial intelligence is transforming the world.",
    "Hallucination in LLMs occurs when the model generates false information."
]

for frase in frases_en:
    tokens = tokenizer.encode(frase)
    palavras = tokenizer.tokenize(frase)
    print(f"\nFrase: {frase}")
    print(f"Nº de tokens: {len(tokens)}")
    print(f"Tokens: {palavras}")
```

```python
frases_pt = [
    "O gato estava sentado no tapete.",
    "A inteligência artificial está a transformar o mundo.",
    "A alucinação em LLMs ocorre quando o modelo gera informação falsa."
]

for frase in frases_pt:
    tokens = tokenizer.encode(frase)
    palavras = tokenizer.tokenize(frase)
    print(f"\nFrase: {frase}")
    print(f"Nº de tokens: {len(tokens)}")
    print(f"Tokens: {palavras}")
```

```python
pares = [
    ("The cat sat on the mat.", "O gato estava sentado no tapete."),
    ("Artificial intelligence", "Inteligência artificial"),
    ("Hallucination detection", "Detecção de alucinações"),
]

print("=" * 60)
print(f"{'Frase':<40} {'Tokens':>6}")
print("=" * 60)

for en, pt in pares:
    n_en = len(tokenizer.encode(en))
    n_pt = len(tokenizer.encode(pt))
    print(f"EN: {en:<38} {n_en:>6}")
    print(f"PT: {pt:<38} {n_pt:>6}")
    print(f"{'Diferença PT-EN:':<38} {n_pt - n_en:>+6}")
    print("-" * 60)
```

```python
palavras_pt = [
    "alucinação",
    "unhappy",
    "descontextualização",
    "fact-checking",
    "infelizmente",
    "hallucination"
]

print("Como o tokenizer divide palavras complexas:\n")
for palavra in palavras_pt:
    tokens = tokenizer.tokenize(palavra)
    print(f"{palavra:<25} → {tokens}")
```

## O que observei sobre tokenização EN vs PT

**Descoberta 1:** O português usa mais tokens que o inglês para frases equivalentes porque s palavras são mais longas, têm mais variações de género, número e conjugação.

**Descoberta 2:** Exemplo da tokenização para palavras longas portuguesas - "descontextualização" - é dividida em 6 partes porque o tokenizador não tem essa palavra inteira no seu vocabulário. Já "hallucination" em inglês parte-se em menos pedaços porque o vocabulário foi optimizado para padrões do inglês.

**Descoberta 3:** Isto importa para AI evaluation porque cada modelo tem uma janela de processamento de tokens, no caso de PT usa-se mais tokens e esta janela fecha. Ou seja, o modelo acaba perdendo informação relevante mais cedo em textos longos em PT; a coerência em respostas longas é mais frágil em PT; o custo da API em tarefa PT é um pouco maior porque usa mais tokens.

**Implicação para o meu trabalho:** Quando avalio outputs em PT, o modelo está a processar mais tokens do que em EN para o mesmo significado. Isto 
pode explicar inconsistências em respostas longas em português que não aparecem nas equivalentes em inglês. Como evaluator bilíngue, consigo 
distinguir um erro linguístico genuíno de uma limitação de contexto — e isso é um diferencial que poucos avaliadores têm.
